```python
import pandas as pd
student1 = pd.read_csv("student1.csv", na_values = [" "])
student2 = pd.read_csv("student2.csv")
student3 = pd.read_csv("student3.csv")
```

#Concatenate data row ie. one dataframe on top of other dataframe

We are concating student1 and student2 to make one dataframe. this concatenation will be vertical. ie. on axis rows or 0

---



```python
student = pd.concat([student1, student2], axis = "rows", ignore_index = True)
student
```

```python
student3
```

#Concatenate data column wise  ie. one dataframe on side of other dataframe

We are concating student and student3 to make one dataframe. this concatenation will be horizontal. ie. on axis columns or 1

---



```python
student = pd.concat([student, student3], axis = "columns")
```

#Merge

##Inner Join

Find details of student and their applications

```python
pd.merge(student, apply, on= "sid")
```

##Left Join

Find details of Student who have not applied yet!!

```python
student_not_applied = pd.merge(student, apply, on = "sid", how = "left")
student_not_applied
```

Check row 5, 10, 11, 13, 14 and 19
These student have not applied anywhere
therefore they have NaN in application columns

```python
# Find rows where any column of application is null
# we took major
# so code will be
student_not_applied[student_not_applied.major.isna()]
```

Now, you have only those rows where student had not applied.

##Right join
We will find application whose corresponding student don't exist!!

```python
application_without_student = pd.merge(student, apply, on = "sid", how = "right")
application_without_student
```

Check row 5 and 6. Application is in the system but no details of student.

```python
application_without_student[application_without_student.sname.isna()]
```

Drop these rows

```python
rows_to_delete = application_without_student[application_without_student.sname.isna()].index
rows_to_delete
```

```python
application_without_student.shape
```

```python
application_without_student.drop(rows_to_delete)
```
