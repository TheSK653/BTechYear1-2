```python
from matplotlib import pyplot as plt
import numpy as np
```

# 1

```python
# Generate data
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

X=np.arange(0,361,10)
Y=[np.sin(np.deg2rad(x)) for x in X]

# Plotting
plt.plot(X, Y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sinusoidal Curve')
plt.grid()
plt.show()
```

```python
### 1. Plotting a Linear Function
# Create a line plot of the linear function ( y = 2x + 3 ) for \( x \) values ranging from -10 to 10
x=np.linspace(-10,10,100)
y=x*2+3
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('2y+3')
plt.title('Plotting a Linear Function: y = 2x + 3, x')
plt.grid()
plt.show()
```

```python
### 2. Plotting Multiple Functions
# Plot the functions \( y = x^2 \) and \( y = \sqrt{x} \) on the same plot for \( x \) values ranging
# from 0 to 10. Use different line styles or colors for each function.
i=np.linspace(0,10,100)
x=i**2
y=np.sqrt(i)
plt.plot(i,x,label='y=x^2')
plt.plot(i,y,label='y=sqrt(x)')
plt.title('Plotting Multiple Functions: ( y = x^2 ) and ( y = sqrt{x} )')
plt.legend()
plt.grid()
plt.show()
```

```python
### 3. Plotting Trigonometric Functions
# Create a line plot showing the functions \( y = \sin(x) \), \( y = \cos(x) \), and \( y = \tan(x) \) for
# \( x \) values ranging from 0 to \( 2\pi \).
i=np.linspace(0,2*np.pi,100)
x=np.sin(i)
y=np.cos(i)
z=np.tan(i)
plt.plot(i,x,label='sin(x)')
plt.plot(i,y,label='cos(x)')
plt.plot(i,z,label='tan(x)')
plt.xlabel('rad')
plt.title('Trigonometric Functions')
plt.ylim(-1,1)
plt.legend()
plt.grid()
plt.show()
```

```python
### 8. Plotting Step Function
# Create a line plot of the step function ( f(x) = begin{cases} 0 & text{if } x < 0 \\ 1 & text{if } x
# geq 0 end{cases} ) for ( x ) values ranging from -5 to 5.
x=np.linspace(-5,5,100)
y=[0 if x<0 else 1 for x in x]
plt.step(x,y,where='mid')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Step funtion')
plt.grid()
plt.show()
```
